Find the

### Mean using basic python

```python
def mean_calc(data_set):
    return sum(data_set)/len(data_set)
```

```python

Total_runs_Conceded_Match = [50,23,10,25,32,12,10,48,27,31,29,24,28,31,26,17,43,24,19,19]
print(f'There are {len(Total_runs_Conceded_Match)} matches of  in the data set')
```

```python
print(mean_calc(Total_runs_Conceded_Match) == np.mean(Total_runs_Conceded_Match))
```

```python
mean_Total_runs_Conceded_Match = mean_calc(Total_runs_Conceded_Match)
mean_Total_runs_Conceded_Match
```

### Median using basic python

```python
def median_calc(data_set):
    sorted_data_set = list(sorted(data_set))
    data_set_length = len(sorted_data_set)
    
    if data_set_length == 1:
        median = data_set[0]
    elif data_set_length % 2 != 0:
        median = sorted_data_set[int((data_set_length)/2)]
    else:
        median = (sorted_data_set[int(data_set_length/2)+1]+sorted_data_set[int(data_set_length/2)])/2
    
    return median
```

```python
median_calc(Total_runs_Conceded_Match) == np.median(Total_runs_Conceded_Match)
median_Total_runs_Conceded_Match = median_calc(Total_runs_Conceded_Match)
median_Total_runs_Conceded_Match
```

### Mode using basic python

```python
def mode_calc(data_set):
    return max(set(data_set), key=data_set.count)

mode_Total_runs_Conceded_Match = mode_calc(Total_runs_Conceded_Match)
mode_Total_runs_Conceded_Match
```

### Plotting mean, median and mode

```python
plt.hist(Total_runs_Conceded_Match,edgecolor='black',color='purple')
plt.axvline(mean_Total_runs_Conceded_Match,color='red',lw=3,label=f'Average (Mean) runs Conceded: ${mean_Total_runs_Conceded_Match:,.0f}')
plt.axvline(median_Total_runs_Conceded_Match,color='orange',lw=3,label=f'Median runs Conceded: ${median_Total_runs_Conceded_Match:,.0f}')
plt.axvline(mode_Total_runs_Conceded_Match,color='blue',lw=3,label=f'Mode runs Conceded: ${mode_Total_runs_Conceded_Match:,.0f}')

plt.xlabel('Total runs Conceded Match')
plt.ylabel('Count')
plt.title('Jasprit Bumraha last 20 matches of T20')
plt.legend();
```

**From our data set, we can see that Jasprit Bumraha total runs conceded in T20 matches in last 20 matches are skewed right (Mode < Median < Mean).**

```python
house_prices = [231000, 210155, 248159, 210007, 224949, 109364, 180758, 115347, 192100, 209048, 159021,
                240542, 205597, 132563, 238747, 245224, 204391, 245126, 262783, 231659, 205425, 203841,
                204779, 240997, 246711, 201422, 209572, 167624, 112952, 266321, 175111, 205689, 205689,
                199416, 243123, 201749, 294213, 238547, 204779, 328371, 147633, 163145, 162602, 156899,
                247632, 256779, 246363, 175603, 245942, 189346, 144480, 174290, 267967, 144447, 271066, 
                221981, 230613, 181986, 202280, 169995, 264285, 236984, 217176, 262220, 226031, 120882,
                158064, 229017, 202858, 309828, 287044, 105005, 198685, 236472, 151793, 174376, 203841,
                207357, 239495, 209001, 194933, 163904, 204779, 222215, 166573, 247718, 406233, 210111,
                215963, 218447, 277012]
```

* **Variance**

```python
'''Without importing library'''
n = len(house_prices)
mean = sum(house_prices)/n
print("The mean of the given data:", mean)
var = sum((item-mean)**2 for item in house_prices)/(n-1)
print("The variance of the given data:", var)
```

```python
'''Using statistics library'''

print("The mean of the given data:",statistics.mean(house_prices))
var = statistics.variance(house_prices)
print("The variance of the given data:", var)
```

* **Standard Deviation**

```python
std = math.sqrt(var)
print("The standard deviation of the given data: ", std)
```

```python
'''Using statistics library'''

std = statistics.stdev(house_prices)
print("The standard deviation of the given data: ", std)
```

#### Plotting mean and standard deviation

```python
plt.hist(house_prices, edgecolor = 'black')
plt.axvline(mean, color = 'purple',lw = 3,label = f'Average (Mean)\nHouse Price: ${mean:,.0f}')
plt.axvline(std, color = 'red',lw = 3,label = f'standard deviation\nHouse Price: ${std:,.0f}')
plt.xlabel('House Price') 

plt.xticks([num for num in range(0,410000,50000)],
          ['$'+str(num) + ',000' if len(str(num)) > 1 else '$' + str(num) for num in range(0,410,50)]) 


plt.yticks(range(0,25,5)) 

plt.ylabel('Count') 
plt.title('House Prices') 
plt.legend(); 
```

A low standard deviation indicates that the values tend to be close to the mean (also called the expected value) of the set, while a high standard deviation indicates that the values are spread out over a wider range.

Find and interperate
* Covariance
* Correlation coefficient

```python
h = [174,189,185,195,149,189,147,154,174,192,155,191,153,157,140,144,172,157,153,169,185,172,151]
w = [96,87,110,104,61,104,92,111,90,101,51,79,107,110,129,145,139,110,149,97,139,67,64]
```

* **Covariance**

```python
n = len(h)
mean_h, mean_w = sum(h)/n, sum(w)/n
cov_hw = (sum((h[i]-mean_h)*(w[i]-mean_w) for i in range(n))/(n-1))
print("The Covariance of height and weight is", cov_hw)
```

* **Correlation Coefficient**

```python
n = len(h)
mean_h, mean_w = sum(h)/n, sum(w)/n
cov_hw = (sum((h[i]-mean_h)*(w[i]-mean_w) for i in range(n))/(n-1))
var_h = sum((item-mean_h)**2 for item in h)/ (n-1)
var_w = sum((item-mean_w)**2 for item in w)/ (n-1)
std_h, std_w = var_h**0.5, var_w**0.5
r = cov_hw/(std_h * std_w)
print("The correlation coefficient of height and weight is", r)
```
