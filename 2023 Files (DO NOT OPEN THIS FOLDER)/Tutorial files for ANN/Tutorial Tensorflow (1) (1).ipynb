## Automatic Differentiation in Tensorflow

Gradient descent and related algorithms are a cornerstone of modern machine learning.

To enable this, TensorFlow implements automatic differentiation (autodiff), which uses calculus to compute gradients. Typically you'll use this to calculate the gradient of a model's error or loss with respect to its weights.

#### Task 12. Differentiate the function y = x**4 + 2*x**3 - 4*x**2 + 2*x -8 at the point x = 5

```python
import tensorflow as tf
x = tf.Variable(5.0)

def f(x):
  y = x**4 + 2*x**3 - 4*x**2 + 2*x -8
  return y
```

```python
with tf.GradientTape() as tape:
  y = f(x)

g_x = tape.gradient(y, x)  # g(x) = dy/dx

g_x
```

## Tensorflow Session (switches to control your dataflow graphs)

So big picture in small words see in any tensorflow programme consists of 2 parts
Data-flow (series of tensorflow operations arranged of graph nodes)
Running computation graph in a session (encapsulates control and tensorflow runtime)

#### TAsk 13. Use the tensorflow version 1 to create a dataflow graph to calculate a*b + c**2 , (a=2,b=5,c=4) and run it in a tensorflow session.

Disable the tensorflow version 2 to do the given task

```python
import tensorflow.compat.v1 as tfo
tfo.disable_v2_behavior()
# Create an empty graph, that will be filled with operation nodes later
main_graph=tfo.Graph()


# Register this graph as default graph. 
# All operations within this context will become operation nodes in this graph
with main_graph.as_default():
    
    # Create two constants of value 5
    a=tfo.constant(2.0)
    b=tfo.constant(5.0)
    c=tfo.constant(4.0)
    
    # Multiply the constants with each other
    mul=tfo.multiply(a,b)
    y=mul + c**2
    
    

# Create a session to execute the dataflow graph
with tfo.Session(graph=main_graph) as session:
    
    # Perform the calculation defined in the dataflow graph and get the result
    output=session.run(y)
    print('Result : %d '%output)
```

#### Task 14. Create a Tensorflow placeholder 'a' and perform the operation 'b= a^3 + a^2'. then create a tensorflow session and find the values of 'b' for 'a'= 1,2,3,4,5 .  

```python
# importing packages
#import tensorflow.compat.v1 as tf
  
# disabling eager mode
#tf.compat.v1.disable_eager_execution()
  
# creating a placeholder
a = tfo.placeholder(tf.float32, None)
  
# creating an operation
b = a**3 + a**2
  
# creating a session
with tfo.Session() as session:
    
    # feeding data in the placeholder
    operation_res = session.run(b, feed_dict={a: [1, 2, 3, 4, 5]})
    print("after executing the operation: " + str(operation_res))
```
