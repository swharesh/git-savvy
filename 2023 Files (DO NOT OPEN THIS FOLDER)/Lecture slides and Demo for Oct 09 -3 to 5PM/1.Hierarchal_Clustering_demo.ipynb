Example with synthetic data:



```python
import numpy as np

X = np.array([[4,2], [9,16], [14,12], [25,11], [29,30], [31,31], [84,72], [73,80], [63,79], [72,51], [82,90], [87, 96]])
```

```python
#Plotting and marking data points:
import matplotlib.pyplot as plt

labels = range(1, 13)
plt.figure(figsize=(6, 6))
plt.subplots_adjust(bottom=0.1)
plt.scatter(X[:,0],X[:,1], label='True Position')

for label, x, y in zip(labels, X[:, 0], X[:, 1]):
    plt.annotate(
        label,
        xy=(x, y), xytext=(-3, 3),
        textcoords='offset points', ha='right', va='bottom')
plt.show()
```

```python
#Dendogram with Single Linkage
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt

linked = linkage(X, 'single')

labelList = range(1, 13)

plt.figure(figsize=(10, 7))
dendrogram(linked,
            orientation='top',
            labels=labelList,
            distance_sort='descending')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean',
                                  linkage='single')
cluster.fit_predict(X)
```

```python
plt.scatter(X[:,0],X[:,1], c=cluster.labels_, cmap='rainbow')
```
